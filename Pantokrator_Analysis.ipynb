- **Йерархическа сложност**: Типът „Христос Велик Архиерей“ демонстрира догматичната зрялост на школата и нейната способност да визуалзира сложни теологични концепции.

### **5. Бъдещо развитие: Computer Vision в услуга на изкуствознанието**

Следващата фаза на това изследване е фокусирана върху технологичната етика и дигиталното съхранение. 
Предложените Computer Vision методи – от дешифриране на надписи през потъмнял лак до дигитална реконструкция на физически унищожени икони – не са просто техническа иновация. Те са единственият инструмент за „възкресяване“ на загубено знание.

Прилагането на алгоритми за подобряване на контраста върху архивни кадри от Самоковската школа ще позволи на изкуствоведа да извърши палеографски анализ на текстовете в Евангелията, които днес са невидими. Това превръща компютърните науки в съюзник на паметта, предотвратявайки окончателното заличаване на творби, които вече не съществуват във физическия свят.

Примерен работен процес (Workflow):

- **Denoising & Contrast Enhancement**: Премахване на шума и засилване на контраста за четене на надписи.

- **Edge Detection**: Откриване на оригиналния контур под повредените слоеве.

- **OCR (Optical Character Recognition)**: Опит за автоматично разпознаване на старославянския текст в разтворената книга.

В рамките на бъдещото развитие на проекта се предвижда прилагането на алгоритъма CLAHE (Contrast Limited Adaptive Histogram Equalization) като ключов инструмент за дигитална реставрация и иконографски анализ.

За разлика от стандартната хистограмна еквилизация, която прилага глобална трансформация върху цялото изображение (често водеща до преекспониране и изкуствен шум), CLAHE оперира на локално ниво. Изображението се разделя на малки контекстуални региони (плочки), във всеки от които се изчислява локален контраст.

Основните предимства на CLAHE в контекста на Самоковската школа са:

Преодоляване на деградацията на лаковото покритие: Алгоритъмът успешно извлича детайли от силно потъмнели и окислени участъци на иконите, където човешкото око вижда само черен фон.

Лимитиране на шума (Contrast Limiting): Чрез поставяне на граница на усилването на контраста, CLAHE предотвратява усилването на микро-пукнатини (кракле) и дефекти в основата, които биха могли да бъдат объркани с фрагменти от надписи.

Палеографска четливост: Този метод е критично важен за визуализирането на микро-текстовете в разтворените книги на Христос, позволявайки прецизно разчитане на цитатите от старославянския канон, дори при частично разрушен живописен слой.

Прилагането на CLAHE превръща повредения дигитален архив от пасивно хранилище в активен източник на нова научна информация.

### Визуален пример: Възстановяване на детайли чрез CLAHE
Поради естественото стареене на лаковете и натрупването на сажди, много от иконите са почти невидими за стандартните сензори. Чрез алгоритъма CLAHE успяваме да извлечем скритите детайли.

```python
from IPython.display import Image, display
display(Image(filename='icon.jpg', width=400))
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Зареждане на иконата
img = cv2.imread('icon.jpg', 0) # Четем я в сивата гама за по-ясен контраст

# Прилагане на CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)

# Визуализация на сравнението
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Оригинал')
plt.subplot(1, 2, 2), plt.imshow(cl1, cmap='gray'), plt.title('След CLAHE обработка')
plt.show()
```
